```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_test=pd.read_csv(r'C:\Users\KeerthyCherry\Downloads\weatherAUS.csv')
df_test.tail(10)
```

```python
#info of data
df_test.info()
```

```python
# shape of data:
print(f'Number of columns: { df_test.shape[1]} and Number of rows: {df_test.shape[0]}')
```

```python
# statistical info of dataset
df_test.describe().T
```

```python
df_test["RainTomorrow"].value_counts().plot(kind="bar", color=["lightblue", "salmon"]);
```

```python
fig, ax = plt.subplots(figsize=(10,7))

ax.scatter(df_test.MaxTemp,
            df_test.Rainfall,
            color=["violet"])
plt.title("MaxTemp vs Rainfall")
plt.ylabel("Rainfall")
plt.xlabel("MaxTemp");
```

```python
fig, ax = plt.subplots(figsize=(10,7))

ax.scatter(df_test.MinTemp,
            df_test.Rainfall,
            color=["green"])
plt.title("MinTemp vs Rainfall")
plt.ylabel("Rainfall")
plt.xlabel("MainTemp");
```

```python
fig, ax = plt.subplots(figsize=(10,6))

ax.scatter(df_test.Date[:1000],
           df_test.Rainfall[:1000],
           color=["blue"])
plt.title("Date vs Rainfall")
plt.ylabel("Rainfall")
plt.xlabel("Date");
```

```python

fig, ax = plt.subplots(figsize=(7,7))

ax.scatter(df_test.Date[:1000],
           df_test.WindGustSpeed[:1000],
           color=["blue"])
plt.title("WindGustSpeed by Date")
plt.ylabel("WindGustSpeed")
plt.xlabel("Date");
```

```python
fig, ax = plt.subplots(figsize=(7,7))

ax.scatter(df_test.Date[:1000],
           df_test.Rainfall[:1000],
           color=["darkred"])
plt.title("Rainfall by Date")
plt.ylabel("Rainfall")
plt.xlabel("Date");
```

```python
fig,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(9,7),
                             sharex=True)

# Scatter plot with WindSpeed9am
ax0.scatter(df_test.Date[:1000],
            df_test.WindSpeed9am[:1000],
            color="teal");

ax0.set(title="WindSpeed9am vs Date",
        xlabel="date",
        ylabel="WindSpeed9am")

# Scatter plot with WindSpeed3pm
ax1.scatter(df_test.Date[:1000],
            df_test.WindSpeed3pm[:1000],
            color="red")

ax1.set(title="WindSpeed3pm vs Date",
        xlabel="date",
        ylabel="WindSpeed3pm");
```

```python
fig,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,8),
                             sharex=True)

# Scatter plot with Humidity9am
ax0.scatter(df_test.Date[:1000],
            df_test.Humidity9am[:1000],
            color="teal");
ax0.set(title="Humidity9am vs Date",
        xlabel="date",
        ylabel="Humidity9am")

# Scatter plot with Humidity3pm
ax1.scatter(df_test.Date[:1000],
            df_test.Humidity3pm[:1000],
            color="blue")

ax1.set(title="Humidity3pm vs Date",
        xlabel="date",
        ylabel="Humidity3pm");

```

```python
#From this we can infer that Humdity is lowest at January and starts increasing from February till August it is at it's peak on July and then from September to December it starts falling
```

```python
fig,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,8),
                             sharex=True)

# Scatter plot with Pressure9am
ax0.scatter(df_test.Date[:1000],
            df_test.Pressure9am[:1000],
            color="green");

ax0.set(title="Pressure9am vs Date",
        xlabel="date",
        ylabel="Pressure9am")

# Scatter plot with Pressure3pm
ax1.scatter(df_test.Date[:1000],
            df_test.Pressure3pm[:1000],
            color="violet")

ax1.set(title="Pressure3pm vs Date",
        xlabel="date",
        ylabel="Pressure3pm");
```

```python
fig,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,8),
                             sharex=True)

# Scatter plot with Temp9am
ax0.scatter(df_test.Date[:1000],
            df_test.Temp9am[:1000],
            color="navy");

ax0.set(title="Temp9am vs Date",
        xlabel="date",
        ylabel="Temp9am")

# Scatter plot with Temp3pm
ax1.scatter(df_test.Date[:1000],
            df_test.Temp3pm[:1000],
            color="firebrick")

ax1.set(title="Temp3pm vs Date",
        xlabel="date",
        ylabel="Temp3pm");
```

```python
#As Australia lies in the Southern Hemisphere the Summer Season is from December to February and Winter is 
from June to August so we can see high temps during summer and low during winter
```

```python
fig,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,8),
                             sharex=True)

# Scatter plot with Maxtemp
ax0.scatter(df_test.Date[:1000],
            df_test.MaxTemp[:1000],
            color="teal");

ax0.set(title="Maxtemp vs Date",
        xlabel="date",
        ylabel="Maxtemp")

# Scatter plot with MinTemp
ax1.scatter(df_test.Date[:1000],
            df_test.MinTemp[:1000],
            color="firebrick")

ax1.set(title="MinTemp vs Date",
        xlabel="date",
        ylabel="MinTemp");
```

```python
#from this we can conclude that During Summer

#maximum MaxTemp is over 45 degrees
#maximum MinTemp is over 25 degrees
#minimum MaxTemp is between 20-30 degrees
#minimum MinTemp is between 5-15 degrees

#During Winter

#maximum MaxTemp is between 20-25 degrees
#maximum MinTemp is between 5-10 degrees
#minimum MaxTemp is just under 10 degrees
#minimum MinTemp is under 0 degrees
```
